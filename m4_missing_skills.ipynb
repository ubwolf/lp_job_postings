```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.feature_extraction.text import TfidfVectorizer

def load_pickle(path, file):
    df = pd.read_pickle(os.path.join(path, file))
    return df

path = './output/'
file_c = 'step3_cluster_labels.pk'
file_t = 'step3_skill_point.pk'
df_cluster = load_pickle(path, file_c) 
df_text = load_pickle(path, file_t)
resume_skills = open('./data/Liveproject Resume.txt', 'r').read()
df_data = df_cluster.join(df_text)
df_data.head()
```

```python
df_w_resume = df_data.append({'skills':resume_skills}, ignore_index=True)

def cosine_similarity(df, col, col_list):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(col_list)
    tfidf_matrix = tfidf_matrix.toarray()
    df_new = df.copy()
    df_new[col] = tfidf_matrix[:-1] @ tfidf_matrix[-1]
    df_new = df_new.sort_values(col, ascending=False)
    return df_new

cosine_sim = cosine_similarity(df_data, 'cosine', df_w_resume['skills'])
```

```python
skills_df = cosine_sim.sort_values(['Cluster', 'cosine'], ascending=False).groupby('Cluster').head(5)
skills_df
```
